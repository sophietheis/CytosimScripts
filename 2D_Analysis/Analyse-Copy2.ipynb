```r
# Calcul de la taille des fibres
x_start_fiber_left = data_fiber_left[seq(3, 34, 5)]
x_end_fiber_left   = data_fiber_left[seq(5, 36, 5)]
y_start_fiber_left = data_fiber_left[seq(4, 35, 5)]
y_end_fiber_left   = data_fiber_left[seq(6, 37, 5)]

x_start_fiber_right = data_fiber_right[seq(3, 34, 5)]
x_end_fiber_right   = data_fiber_right[seq(5, 36, 5)]
y_start_fiber_right = data_fiber_right[seq(4, 35, 5)]
y_end_fiber_right   = data_fiber_right[seq(6, 37, 5)]

#size_fibres_left  = (sqrt((data_fiber_left[x_end_fiber]-data_fiber_left[x_start_fiber])^2+(data_fiber_left[y_start_fiber]-data_fiber_left[y_end_fiber])^2))
#size_fibres_rigth = (sqrt((data_fiber_right[x_end_fiber]-data_fiber_right[x_start_fiber])^2+(data_fiber_right[y_start_fiber]-data_fiber_right[y_end_fiber])^2))


size_fibres_left_1 = sqrt((x_start_fiber_left$x - x_end_fiber_left$x.1)^2 + (y_start_fiber_left$y - y_end_fiber_left$y.1)^2)
size_fibres_left_2 = sqrt((x_start_fiber_left$x.2 - x_end_fiber_left$x.3)^2 + (y_start_fiber_left$y.2 - y_end_fiber_left$y.3)^2)
size_fibres_left_3 = sqrt((x_start_fiber_left$x.4 - x_end_fiber_left$x.5)^2 + (y_start_fiber_left$y.4 - y_end_fiber_left$y.5)^2)
size_fibres_left_4 = sqrt((x_start_fiber_left$x.6 - x_end_fiber_left$x.7)^2 + (y_start_fiber_left$y.6 - y_end_fiber_left$y.7)^2)


size_fibres_right_1 = sqrt((x_start_fiber_right$x - x_end_fiber_right$x.1)^2 + (y_start_fiber_right$y - y_end_fiber_right$y.1)^2)
size_fibres_right_2 = sqrt((x_start_fiber_right$x.2 - x_end_fiber_right$x.3)^2 + (y_start_fiber_right$y.2 - y_end_fiber_right$y.3)^2)
size_fibres_right_3 = sqrt((x_start_fiber_right$x.4 - x_end_fiber_right$x.5)^2 + (y_start_fiber_right$y.4 - y_end_fiber_right$y.5)^2)
size_fibres_right_4 = sqrt((x_start_fiber_right$x.6 - x_end_fiber_right$x.7)^2 + (y_start_fiber_right$y.6 - y_end_fiber_right$y.7)^2)


# Remplacement des NA par 0
#size_fibres_left  = replace(size_fibres_left,  is.na(size_fibres_left),0)
#size_fibres_rigth = replace(size_fibres_rigth, is.na(size_fibres_rigth),0)
```

```r
### Etat des MT de gauche par rapport à ceux de droite
diff_fibre_left_1 = size_fibres_left_1 [2:(length(size_fibres_left_1))] - size_fibres_left_1[1:(length(size_fibres_left_1)-1)] 
diff_fibre_left_2 = size_fibres_left_2 [2:(length(size_fibres_left_2))] - size_fibres_left_2[1:(length(size_fibres_left_2)-1)] 
diff_fibre_left_3 = size_fibres_left_3 [2:(length(size_fibres_left_3))] - size_fibres_left_3[1:(length(size_fibres_left_3)-1)] 
diff_fibre_left_4 = size_fibres_left_4 [2:(length(size_fibres_left_4))] - size_fibres_left_4[1:(length(size_fibres_left_4)-1)] 

diff_fibre_left_1[which(diff_fibre_left_1>0)] = 1
diff_fibre_left_1[which(diff_fibre_left_1<0)] = 0
diff_fibre_left_2[which(diff_fibre_left_2>0)] = 1
diff_fibre_left_2[which(diff_fibre_left_2<0)] = 0
diff_fibre_left_3[which(diff_fibre_left_3>0)] = 1
diff_fibre_left_3[which(diff_fibre_left_3<0)] = 0
diff_fibre_left_4[which(diff_fibre_left_4>0)] = 1
diff_fibre_left_4[which(diff_fibre_left_4<0)] = 0

nb_fibres_left_poly = diff_fibre_left_1+diff_fibre_left_2+diff_fibre_left_3+diff_fibre_left_4

diff_fibre_right_1 = size_fibres_right_1 [2:(length(size_fibres_right_1))] - size_fibres_right_1[1:(length(size_fibres_right_1)-1)] 
diff_fibre_right_2 = size_fibres_right_2 [2:(length(size_fibres_right_2))] - size_fibres_right_2[1:(length(size_fibres_right_2)-1)] 
diff_fibre_right_3 = size_fibres_right_3 [2:(length(size_fibres_right_3))] - size_fibres_right_3[1:(length(size_fibres_right_3)-1)] 
diff_fibre_right_4 = size_fibres_right_4 [2:(length(size_fibres_right_4))] - size_fibres_right_4[1:(length(size_fibres_right_4)-1)]


diff_fibre_right_1[which(diff_fibre_right_1>0)] = 1
diff_fibre_right_1[which(diff_fibre_right_1<0)] = 0
diff_fibre_right_2[which(diff_fibre_right_2>0)] = 1
diff_fibre_right_2[which(diff_fibre_right_2<0)] = 0
diff_fibre_right_3[which(diff_fibre_right_3>0)] = 1
diff_fibre_right_3[which(diff_fibre_right_3<0)] = 0
diff_fibre_right_4[which(diff_fibre_right_4>0)] = 1
diff_fibre_right_4[which(diff_fibre_right_4<0)] = 0

nb_fibres_right_poly = diff_fibre_right_1+diff_fibre_right_2+diff_fibre_right_3+diff_fibre_right_4

t0 = 199/0.05
t1 = 201/0.05
#t0 = 500
#t1 = 1000
par(mfrow=c(2,1))
plot (data_solid$time[t0:t1], data_solid$x[t0:t1],type='l', ylim=c(-0.8,0.7), xlab="Time(s)", ylab="Position(µm)")
lines (data_solid$time[t0:t1],data_solid$x.1[t0:t1])
lines (data_solid$time[t0:t1],data_solid$x.2[t0:t1], col="red")
lines (data_solid$time[t0:t1],data_solid$x.3[t0:t1], col="blue")
lines (data_solid$time[t0:t1],data_solid$x[t0:t1]-(data_solid$x[t0:t1]-data_solid$x.1[t0:t1])/2, col = "gray")
lines (data_solid$time[t0:t1],(data_solid$x.2[t0:t1]+data_solid$x.3[t0:t1])/2, col="purple")
plot (data_solid$time[t0:t1],nb_fibres_left_poly[t0:t1], type='l', col="red", ylim=c(0,4), xlab="Time(s)", ylab="Nb polymerising MT")
lines(data_solid$time[t0:t1],nb_fibres_right_poly[t0:t1], col= "blue")

#plot (data_solid$time[t0:t1],nb_fibres_left_poly[t0:t1], type='b',pch=3, col="red", ylim=c(0,4))
#lines(data_solid$time[t0:t1],nb_fibres_right_poly[t0:t1],  type='b',pch=3,col= "blue")
#plot (data_solid$time[t0:t1], data_ktmt_left$force[t0:t1], type = 'b', col = "red")
#lines(data_solid$time[t0:t1], data_ktmt_right$force[t0:t1], type = 'b', col = "blue")
```
