```python
for force in range (0, 200, 10) :
    file_name = "SPAM_"+str(force)+"F.cym"

    file = open(file_name, "w")

    file.write("set simul pombe \n")
    file.write("{\n")
    file.write("    time_step = 0.05\n")
    file.write("    kT         = 0.0042\n")
    file.write("    viscosity  = 0.9\n")
    file.write("}\n")

    file.write("set space pombe\n")
    file.write("{\n")
    file.write("    geometry = ( square 7 )\n")
    file.write("    display = ( color=0xFFFFFF55; )\n")
    file.write("}\n")
    file.write("new space pombe\n")

    file.write("%% --------------------------------------------------------------------------------------------------------------------------------- %%\n")
    file.write("%% ---------------------------------------------------------------SPB--------------------------------------------------------------- %%\n")
    file.write("%% --------------------------------------------------------------------------------------------------------------------------------- %%\n")


    file.write("% Definition des SPBs\n")
    file.write("set fiber microtubule_interdigite_spb1\n")
    file.write("{\n")
    file.write("    confine = inside, 100 \n")
    file.write("    rigidity = 22\n")
    file.write("    segmentation = 0.1\n")

    file.write("    activity         = classic\n")
    file.write("    growing_speed    = 0.03, -0.001\n")
    file.write("    shrinking_speed  = -0.16\n")
    file.write("    catastrophe_rate = 0.002 , 0.02\n")
    file.write("    rescue_rate      = 0.007\n")
    file.write("    growing_force    = 0.01\n")
    file.write("    binding_key = 1\n")
    file.write("    display = {color=light_gray; line_width=0.5}    \n")
    file.write("}\n")

    file.write("set solid spb_corpse1\n")
    file.write("{\n")
    file.write("    steric = 1\n")
    file.write("    confine = inside, 100\n")
    file.write("    external_force = ("+str(force)+")\n")
    file.write("    display = ( color=light_gray; style=2 )\n")
    file.write("}\n")
    file.write("set aster SPB1\n")
    file.write("{\n")
    file.write("    solid = spb_corpse1\n")
    file.write("    fibers = microtubule_interdigite_spb1\n")
    file.write("    nucleation_rate = 0.1\n")
    file.write("    stiffness = 1000, 1000\n")
    file.write("}\n")

    file.write("  new 1 aster SPB1\n")
    file.write("{\n")
    file.write("    position = ( -2.5 0 )\n")
    file.write("    radius = 0.1\n")

    file.write("    %microtubule_interdigite\n")
    file.write("    nb_fibers = 25\n")
    file.write("    length = 4\n")
    file.write("    random_position = straight_aleatory\n")
    file.write("    random_direction = positive\n")
    file.write("    random_pos_x = 0, 0\n")
    file.write("    random_pos_y = -0.01, 0.01\n")
    file.write("    random_pos_z = 0, 0\n")

    file.write("    orientation = identity\n")
    file.write("}\n")

    file.write("set fiber microtubule_interdigite_spb2\n")
    file.write("{\n")
    file.write("    confine = inside, 100 \n")
    file.write("    rigidity = 22\n")
    file.write("    segmentation = 0.1\n")

    file.write("    activity         = classic\n")
    file.write("    growing_speed    = 0.03, -0.001\n")
    file.write("    shrinking_speed  = -0.16\n")
    file.write("    catastrophe_rate = 0.002, 0.02\n")
    file.write("    rescue_rate      = 0.007\n")
    file.write("    growing_force    = 0.01\n")
    file.write("    binding_key = 2\n")
    file.write("    display = {color=light_gray; line_width=0.5}    \n")
    file.write("}\n")

    file.write("set solid spb_corpse2\n")
    file.write("{\n")
    file.write("    steric = 1\n")
    file.write("    confine = inside, 100\n")
    file.write("    external_force = (-"+str(force)+")\n")
    file.write("    display = ( color=light_gray; style=2 )\n")
    file.write("}\n")
    file.write("set aster SPB2\n")
    file.write("{\n")
    file.write("    solid = spb_corpse2\n")
    file.write("    fibers = microtubule_interdigite_spb2\n")
    file.write("    nucleation_rate = 0.1\n")
    file.write("    stiffness = 1000, 1000\n")
    file.write("}\n")
    
    file.write("  new 1 aster SPB2\n")
    file.write("{\n")
    file.write("    position = ( 2.5 0 )\n")
    file.write("    radius = 0.1\n")

    file.write("    %microtubule_interdigite\n")
    file.write("    nb_fibers = 25\n")
    file.write("    length = 4\n")
    file.write("    random_position = straight_aleatory\n")
    file.write("    random_direction = negative\n")
    file.write("    random_pos_x = 0, 0\n")
    file.write("    random_pos_y = -0.01, 0.01\n")
    file.write("    random_pos_z = 0, 0\n")

    file.write("    orientation = identity\n")
    file.write("}\n")
   

    file.write("%% --------------------------------------------------------------------------------------------------------------------------------- %%\n")
    file.write("%% -------------------------------------------------------------MOTEURS------------------------------------------------------------- %%\n")
    file.write("%% --------------------------------------------------------------------------------------------------------------------------------- %%\n")

    file.write("% Definition de KLP9\n")
    file.write("set hand motor1\n")
    file.write("{\n")
    file.write("    binding_rate = 6\n")
    file.write("    binding_range = 0.03\n")
    file.write("    unbinding_rate = 0.1\n")
    file.write("    unbinding_force = 15\n")

    file.write("    activity = move\n")
    file.write("    max_speed = 0.009\n")
    file.write("    stall_force = 5\n")

    file.write("    binding_key = 1\n")
    file.write("    display = { color=green; size=1 }\n")
    file.write("}\n")
    file.write("set hand motor2\n")
    file.write("{\n")
    file.write("    binding_rate = 6\n")
    file.write("    binding_range = 0.03\n")
    file.write("    unbinding_rate = 0.1\n")
    file.write("    unbinding_force = 15\n")

    file.write("    activity = move\n")
    file.write("    max_speed = 0.009\n")
    file.write("    stall_force = 5 % Force pour laquelle la vitesse devient nulle.\n")

    file.write("    binding_key = 2\n")
    file.write("    display = { color=green; size=1 }\n")
    file.write("}\n")
    file.write("%% Definition de couple\n")
    file.write("set couple klp9\n")
    file.write("{\n")
    file.write("    hand1 = motor1\n")
    file.write("    hand2 = motor2\n")

    file.write("    confine = inside, 100\n")
    file.write("    stiffness = 75		% Rigidite du lien\n")
    file.write("    diffusion = 10		% Coefficient de diffusion pour les hands libres\n")
    file.write("    fast_diffusion = 1\n")
    file.write("    length = 0		% Longueur du lien au repos\n")

    file.write("    activity = crosslink\n")
    file.write("    specificity = 1 % BIND_ANTIPARALLEL\n")
    file.write("}\n")

    file.write("% Definition de ASE1\n")
    file.write("set hand ase1_bind1\n")
    file.write("{\n")
    file.write("    binding_rate = 50\n")
    file.write("    binding_range = 0.07\n")
    file.write("    unbinding_rate = 1.8\n")
    file.write("    unbinding_force = 4\n")

    file.write("    binding_key = 1\n")
    file.write("    display = { color=red; size=1 }\n")
    file.write("}\n")
    file.write("set hand ase1_bind2\n")
    file.write("{\n")
    file.write("    binding_rate = 50\n")
    file.write("    binding_range = 0.07\n")
    file.write("    unbinding_rate = 1.8\n")
    file.write("    unbinding_force = 4\n")

    file.write("    binding_key = 2\n")
    file.write("    display = { color=red; size=1 }\n")
    file.write("}\n")
    file.write("set couple ASE1\n")
    file.write("{\n")
    file.write("    hand1 = ase1_bind1\n")
    file.write("    hand2 = ase1_bind2\n")

    file.write("    confine = inside, 100\n")
    file.write("    stiffness = 50		% Rigidite du lien\n")
    file.write("    diffusion = 20		% Coefficient de diffusion pour les hands libres\n")
    file.write("    fast_diffusion = 1\n")
    file.write("    length = 0		% Longueur du lien au repos\n")

    file.write("    activity = crosslink\n")
    file.write("    specificity = 1 % BIND_ANTIPARALLEL\n")
    file.write("}\n")

    file.write("new 500 couple klp9\n")
    file.write("new 500 couple ASE1\n")

    file.write("run simul *\n")
    file.write("{\n")
    file.write("    write_object = solid couple:klp9 couple:ASE1\n")
    file.write("    nb_steps   = 8400\n")
    file.write("    nb_frames  = 8400\n")
    file.write("}\n")





    file.close()
```
